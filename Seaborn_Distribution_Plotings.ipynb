```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd

sns.set_theme(style='ticks', context='talk', palette='deep')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.labelsize'] = 12
```

# Distribution Plots

```python
# Importing Data

data = pd.read_csv("Fashion_Retail_Sales.csv")
df = pd.DataFrame(data)
print(df)
```

```python
# Histogram
fig, ax = plt.subplots()
sns.histplot(data=df, x='Purchase Amount (USD)', bins=24, kde=True, color='#4C72B0', ax=ax)
ax.set_title('Distribution of Purchase Amount (USD)')
ax.set_xlabel('Purchase Amount (USD)')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# KDE
fig, ax = plt.subplots()
sns.kdeplot(data=df, x='Purchase Amount (USD)', fill=True, color='#55A868', ax=ax)
ax.set_title('KDE of Purchase Amount (USD)')
ax.set_xlabel('Purchase Amount (USD)')
ax.set_ylabel('Density')
plt.tight_layout()
plt.show()
```

```python
# Displot
g = sns.displot(data=df, x='Purchase Amount (USD)', kde=True, bins=24, height=5, aspect=1.8, color='#C44E52')
g.fig.suptitle('Purchase Amount Distribution with KDE', y=1.02)
g.set_axis_labels('Purchase Amount (USD)', 'Count')
plt.show()
```

```python
# Boxplot
fig, ax = plt.subplots()
sns.boxplot(y=df['Review Rating'], color='#8172B2', ax=ax)
ax.set_title('Review Rating Distribution')
ax.set_xlabel('')
ax.set_ylabel('Review Rating')
plt.tight_layout()
plt.show()
```

```python
# Violin Plot
fig, ax = plt.subplots()
sns.violinplot(x='Purchase Amount (USD)', y='Review Rating', data=df, color='#64B5CD', ax=ax)
ax.set_title('Review Rating vs Purchase Amount')
ax.set_xlabel('Purchase Amount (USD)')
ax.set_ylabel('Review Rating')
plt.tight_layout()
plt.show()
```

```python
# RugPlot
fig, ax = plt.subplots()
sns.rugplot(x=df['Purchase Amount (USD)'], color='#4C72B0', height=0.08, ax=ax)
ax.set_title('Rug Plot of Purchase Amount (USD)')
ax.set_xlabel('Purchase Amount (USD)')
ax.set_ylabel('')
plt.tight_layout()
plt.show()
```

```python
# ECDF
fig, ax = plt.subplots()
sns.ecdfplot(data=df, x='Purchase Amount (USD)', color='#C44E52', ax=ax)
ax.set_title('ECDF of Purchase Amount (USD)')
ax.set_xlabel('Purchase Amount (USD)')
ax.set_ylabel('Cumulative Probability')
plt.tight_layout()
plt.show()
```
