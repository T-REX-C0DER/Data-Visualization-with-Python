```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# MATPLOTLIB

```python
# Sample data

product_sales = {
    "Product": ["Laptop", "Phone", "Tablet", "Watch"],
    "Online_Sales": [120, 200, 90, 60],
    "Offline_Sales": [80, 150, 60, 40]
}

df = pd.DataFrame(product_sales)

print(df)
```

Basic Methods

```python
plt.plot(product_sales["Online_Sales"] , product_sales["Offline_Sales"])
```

```python
plt.scatter(product_sales["Online_Sales"] , product_sales["Offline_Sales"])
```

```python
plt.bar(product_sales["Online_Sales"] , product_sales["Offline_Sales"])
```

```python
plt.hist(product_sales["Online_Sales"])
```

```python
plt.pie(product_sales["Online_Sales"] , labels = product_sales["Product"])
```

Important Styling Methods

```python
plt.scatter(product_sales["Online_Sales"] , product_sales["Offline_Sales"])
plt.title("Sales Data")
plt.xlabel("Offline Sales")
plt.ylabel("Online Sales")
plt.grid(True)
plt.Figure(figsize=(8,5))
```
