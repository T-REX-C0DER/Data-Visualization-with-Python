```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# MATPLOTLIB

```python
# Importing the sales data

product_sales = {
    "Product": ["Laptop", "Phone", "Tablet", "Watch"],
    "Online_Sales": [120, 200, 90, 60],
    "Offline_Sales": [80, 150, 60, 40]
}

df = pd.DataFrame(product_sales)

print(df)
```

# Basic Methods

```python
plt.plot(product_sales["Online_Sales"] , product_sales["Offline_Sales"])
```

```python
plt.scatter(product_sales["Online_Sales"] , product_sales["Offline_Sales"])
```

```python
plt.bar(product_sales["Online_Sales"] , product_sales["Offline_Sales"])
```

```python
plt.hist(product_sales["Online_Sales"])
```

```python
plt.pie(product_sales["Offline_Sales"] , labels = product_sales["Product"])
```

```python
plt.boxplot(product_sales["Online_Sales"], product_sales["Offline_Sales"])
```

# Important Styling Methods

```python
plt.scatter(product_sales["Online_Sales"] , product_sales["Offline_Sales"])
plt.title("Sales Data")
plt.xlabel("Offline Sales")
plt.ylabel("Online Sales")
plt.grid(True)
plt.Figure(figsize=(8,5))
```

# Subplots

```python
# fig = complete box , ax = every single box 
fig, ax = plt.subplots(2,2, figsize=(10,6))
fig.suptitle("Product Sales Dashboard")

# First section
ax[0,0].bar(product_sales["Product"], product_sales["Online_Sales"])
ax[0,0].set_title("Online Sales")
ax[0,0].set_xlabel("Product")
ax[0,0].set_ylabel("Units Sold")

# Second section
ax[0,1].bar(product_sales["Product"], product_sales["Offline_Sales"])
ax[0,1].set_title("Offline Sales")

# Third section 
ax[1,0].pie(product_sales["Offline_Sales"] , labels = product_sales["Product"])
ax[1,0].set_title("Sales Data")

# Fourth section
ax[1,1].plot(product_sales["Product"], product_sales["Online_Sales"], label="Online")
ax[1,1].plot(product_sales["Product"], product_sales["Offline_Sales"], label="Offline")
ax[1,1].set_title("Online vs Offline")
ax[1,1].legend()

plt.tight_layout()
plt.show()

```

```python
# Saving the figures 
fig.savefig("Grapgh.jpeg" , dpi = 200)
```
