```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

# MATPLOTLIB

```python
# Importing the sales data

sales_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
              "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Sales_2024": [120, 135, 150, 170, 160, 180, 200, 210, 190, 220, 230, 250],
    "Sales_2025": [130, 140, 160, 175, 165, 190, 210, 220, 200, 230, 240, 260]
}

df = pd.DataFrame(sales_data)
print(df)
```

# Basic Methods

```python
plt.plot(df["Sales_2025"] , df["Sales_2024"])
plt.title("Item Sales")
plt.xlabel("Item Purchased")
plt.ylabel("Purchase Amount")
```

```python
plt.scatter(df["Sales_2025"] , df["Sales_2024"])
plt.title("Item Sales")
plt.xlabel("Item Purchased")
plt.ylabel("Purchase Amount")
```

```python
plt.bar(df["Sales_2025"] , df["Sales_2024"])
plt.title("Item Sales")
plt.xlabel("Item Purchased")
plt.ylabel("Purchase Amount")
```

```python
plt.hist(sales_data["Payment Method"])
plt.title("Item Sales")
plt.xlabel("Item Purchased")
plt.ylabel("Purchase Amount")
```

```python
plt.pie(df["Sales_2025"] , labels = df["Month"])
plt.title("Item Sales")
```

# Important Styling Methods

```python
plt.scatter(df["Sales_2025"] , df["Sales_2024"])
plt.title("Item Sales")
plt.xlabel("Item Purchased")
plt.ylabel("Purchase Amount")
plt.grid(True)
plt.Figure(figsize=(8,5))
```

# Subplots

```python
# fig = complete box , ax = every single box 


plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.plot(df["Month"], df["Sales_2024"], marker='o')
plt.title("Sales 2024")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.subplot(1,2,2)
plt.plot(df["Month"], df["Sales_2025"], marker='o')
plt.title("Sales 2025")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.tight_layout()
plt.show()

```
