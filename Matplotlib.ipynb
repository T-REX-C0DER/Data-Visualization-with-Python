```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 5)
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.labelsize'] = 11
```

# MATPLOTLIB

```python
# Importing the sales data

sales_data = {
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
              "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    "Sales_2024": [120, 135, 150, 170, 160, 180, 200, 210, 190, 220, 230, 250],
    "Sales_2025": [130, 140, 160, 175, 165, 190, 210, 220, 200, 230, 240, 260]
}

df = pd.DataFrame(sales_data)
print(df)
```

# Basic Methods

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Sales_2024'], marker='o', linewidth=2.2, label='Sales 2024')
plt.plot(df['Month'], df['Sales_2025'], marker='o', linewidth=2.2, label='Sales 2025')
plt.title('Monthly Sales Trend: 2024 vs 2025')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Sales_2024'], df['Sales_2025'], s=70, alpha=0.85, color='#4C72B0', edgecolors='black')
plt.title('Sales Correlation: 2024 vs 2025')
plt.xlabel('Sales 2024')
plt.ylabel('Sales 2025')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
x = np.arange(len(df['Month']))
w = 0.38
plt.bar(x - w/2, df['Sales_2024'], width=w, label='Sales 2024', color='#55A868')
plt.bar(x + w/2, df['Sales_2025'], width=w, label='Sales 2025', color='#C44E52')
plt.xticks(x, df['Month'])
plt.title('Monthly Sales Comparison')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Sales_2025'], bins=8, color='#8172B2', edgecolor='white')
plt.title('Distribution of 2025 Sales Values')
plt.xlabel('Sales 2025')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.pie(df['Sales_2025'], labels=df['Month'], autopct='%1.1f%%', startangle=90, pctdistance=0.8)
plt.title('2025 Sales Share by Month')
plt.tight_layout()
plt.show()
```

# Important Styling Methods

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Sales_2024'], df['Sales_2025'], s=70, alpha=0.85, color='#64B5CD', edgecolors='black')
plt.title('Sales Correlation with Grid')
plt.xlabel('Sales 2024')
plt.ylabel('Sales 2025')
plt.grid(True, alpha=0.35)
plt.tight_layout()
plt.show()
```

# Subplots

```python
# fig = complete canvas, ax = each subplot
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

axes[0].plot(df['Month'], df['Sales_2024'], marker='o', linewidth=2.2, color='#4C72B0')
axes[0].set_title('Sales 2024')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Sales')

axes[1].plot(df['Month'], df['Sales_2025'], marker='o', linewidth=2.2, color='#C44E52')
axes[1].set_title('Sales 2025')
axes[1].set_xlabel('Month')

fig.suptitle('Monthly Sales by Year', y=1.02)
fig.tight_layout()
plt.show()
```

```python
# Saving the figure from previous cell
fig.savefig('Graph.jpeg', dpi=200, bbox_inches='tight')
```
