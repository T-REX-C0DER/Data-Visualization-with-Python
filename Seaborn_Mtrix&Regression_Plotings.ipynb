```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', context='talk', palette='crest')
plt.rcParams['figure.figsize'] = (11, 6)
plt.rcParams['axes.titleweight'] = 'bold'
```

# Matrix Plotings

```python
sales_data = pd.read_csv("Fashion_Retail_Sales.csv")
df = pd.DataFrame(sales_data)
print(df)
```

```python
# Correlation Heat Map
numeric_df = df.select_dtypes(include='number')
corr = numeric_df.corr(numeric_only=True)

plt.figure(figsize=(10, 7))
ax = sns.heatmap(corr, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=0.5)
ax.set_title('Correlation Matrix of Numeric Features')
plt.tight_layout()
plt.show()

# Regression Plot
fig, ax = plt.subplots(figsize=(10, 6))
sns.regplot(data=df, x='Purchase Amount (USD)', y='Review Rating', scatter_kws={'alpha': 0.5, 's': 30}, line_kws={'color': '#C44E52', 'linewidth': 2}, ax=ax)
ax.set_title('Review Rating vs Purchase Amount (Regression)')
ax.set_xlabel('Purchase Amount (USD)')
ax.set_ylabel('Review Rating')
plt.tight_layout()
plt.show()
```
