```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

# Distribution Plots

```python
# Importing Data

data = pd.read_csv("Fashion_Retail_Sales.csv")
df = pd.DataFrame(data)
print(df)
```

```python
# Histogram
plt.figure(figsize=(14,6))
sns.histplot(data=df , x = 'Purchase Amount (USD)' , bins=20 , kde=True)
plt.xticks(rotation=90)  
plt.tight_layout()        
plt.show()
```

```python
# KDE
plt.figure(figsize=(14,6))
sns.kdeplot(data=df , x = 'Purchase Amount (USD)', fill=True)
plt.xticks(rotation=90)  
plt.tight_layout()        
plt.show()
```

```python
# Displot
plt.figure(figsize=(14,6))
sns.displot(data=df , x = 'Purchase Amount (USD)', kde=True)
plt.xticks(rotation=90)  
plt.tight_layout()        
plt.show()
```

```python
# boxplot
plt.figure(figsize=(14,6))
sns.boxplot(y = df["Review Rating"])
plt.xticks(rotation=90)  
plt.tight_layout()        
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(14,6))
sns.violinplot( x = 'Purchase Amount (USD)', y = 'Review Rating' , data=df)
plt.xticks(rotation=90)  
plt.tight_layout()        
plt.show()
```

```python
# RugPlot
plt.figure(figsize=(14,6))
sns.rugplot( x = df["Customer Reference ID"])
plt.xticks(rotation=90)  
plt.tight_layout()        
plt.show()
```

```python
# ECDF
plt.figure(figsize=(14,6))
sns.violinplot( x = 'Purchase Amount (USD)',  data=df)
plt.xticks(rotation=90)  
plt.tight_layout()        
plt.show()
```
