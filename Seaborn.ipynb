```python
import matplotlib.pyplot as plt
import seaborn as sns 
import pandas as pd
```

# Distribution Plots

```python
# Importing Data

data = pd.read_csv("Fashion_Retail_Sales.csv")
df = pd.DataFrame(data)
print(df)
```

```python
# Histogram
plt.figure(figsize=(14,6))
sns.histplot(data=df , x = 'Item Purchased' , bins=20 , kde=True)
plt.xticks(rotation=90)  
plt.tight_layout()        
plt.show()
```
