```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Categorical Plotings 

```python
sales_data = pd.read_csv("Fashion_Retail_Sales.csv")
df = pd.DataFrame(sales_data)
print(df)
```

```python
# CountPlot
sns.countplot(data=df , x = "Item Purchased")
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```
