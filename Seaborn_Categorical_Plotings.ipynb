```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', context='talk', palette='Set2')
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['axes.titleweight'] = 'bold'
plt.rcParams['axes.labelsize'] = 12
plt.rcParams['xtick.labelsize'] = 10
plt.rcParams['ytick.labelsize'] = 10
```

# Categorical Plotings 

```python
sales_data = pd.read_csv("Fashion_Retail_Sales.csv")
df = pd.DataFrame(sales_data)
print(df)
```

```python
df.head()
```

```python
# CountPlot: item purchase frequency
ax = sns.countplot(data=df, x='Item Purchased', order=df['Item Purchased'].value_counts().index, color='#4C72B0')
ax.set_title('Purchase Frequency by Item')
ax.set_xlabel('Item Purchased')
ax.set_ylabel('Count')
plt.xticks(rotation=65, ha='right')
plt.tight_layout()
plt.show()
```

```python
# BarPlot: average purchase amount by item
ax = sns.barplot(data=df, x='Item Purchased', y='Purchase Amount (USD)', estimator='mean', errorbar=None, color='#55A868')
ax.set_title('Average Purchase Amount by Item')
ax.set_xlabel('Item Purchased')
ax.set_ylabel('Average Purchase Amount (USD)')
plt.xticks(rotation=65, ha='right')
plt.tight_layout()
plt.show()
```

```python
# BoxPlot: review rating spread per item
ax = sns.boxplot(data=df, x='Item Purchased', y='Review Rating', color='#C44E52')
ax.set_title('Review Rating Distribution by Item')
ax.set_xlabel('Item Purchased')
ax.set_ylabel('Review Rating')
plt.xticks(rotation=65, ha='right')
plt.tight_layout()
plt.show()
```

```python
# PointPlot: average purchase amount by payment method
ax = sns.pointplot(data=df, x='Payment Method', y='Purchase Amount (USD)', estimator='mean', errorbar=('ci', 95), color='#8172B2')
ax.set_title('Average Purchase Amount by Payment Method')
ax.set_xlabel('Payment Method')
ax.set_ylabel('Purchase Amount (USD)')
plt.tight_layout()
plt.show()
```

```python
# CatPlot: purchase amount by payment method
g = sns.catplot(data=df, x='Payment Method', y='Purchase Amount (USD)', kind='box', height=5, aspect=1.8)
g.fig.suptitle('Purchase Amount Distribution by Payment Method', y=1.02)
g.set_axis_labels('Payment Method', 'Purchase Amount (USD)')
plt.show()
```

```python
# ViolinPlot: purchase amount distribution by payment method
ax = sns.violinplot(data=df, x='Payment Method', y='Purchase Amount (USD)', inner='quartile')
ax.set_title('Purchase Amount Distribution by Payment Method')
ax.set_xlabel('Payment Method')
ax.set_ylabel('Purchase Amount (USD)')
plt.tight_layout()
plt.show()
```

```python
# CatPlot: top 12 most purchased items by payment method
top_items = df['Item Purchased'].value_counts().head(12).index
subset = df[df['Item Purchased'].isin(top_items)]
g = sns.catplot(data=subset, x='Payment Method', hue='Item Purchased', kind='count', height=6, aspect=1.9)
g.fig.suptitle('Top Purchased Items by Payment Method', y=1.02)
g.set_axis_labels('Payment Method', 'Count')
g._legend.set_title('Item Purchased')
plt.show()
```

```python
# StripPlot: review ratings by payment method
ax = sns.stripplot(data=df, x='Payment Method', y='Review Rating', jitter=0.25, alpha=0.65, color='#64B5CD')
ax.set_title('Review Ratings by Payment Method')
ax.set_xlabel('Payment Method')
ax.set_ylabel('Review Rating')
plt.tight_layout()
plt.show()
```

```python
# SwarmPlot: purchase amount by payment method
sample = df.sample(min(500, len(df)), random_state=42)
ax = sns.swarmplot(data=sample, x='Payment Method', y='Purchase Amount (USD)', size=3.5)
ax.set_title('Purchase Amount by Payment Method (Sampled for Readability)')
ax.set_xlabel('Payment Method')
ax.set_ylabel('Purchase Amount (USD)')
plt.tight_layout()
plt.show()
```
