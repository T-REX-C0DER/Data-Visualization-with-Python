```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Categorical Plotings 

```python
sales_data = pd.read_csv("Fashion_Retail_Sales.csv")
df = pd.DataFrame(sales_data)
print(df)
```

```python
df.head()
```

```python
# CountPlot
sns.countplot(data=df , x = "Item Purchased")
plt.xticks(rotation=90)  
plt.tight_layout()
```

```python
# BarPlot
sns.barplot(data=df , x = "Item Purchased", y = "Purchase Amount (USD)")
plt.xticks(rotation=90)  
plt.tight_layout()
```

```python
# BoxPlot
sns.boxplot(data=df , x = "Item Purchased" , y ="Review Rating" , color='pink')
plt.xticks(rotation=90)  
plt.tight_layout()
```

```python
# PointPlot
sns.pointplot(data=df , x = "Item Purchased" , y = "Payment Method" , color='red')
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```

```python
# CatPlot
sns.catplot(data=df , x = "Payment Method" , y = "Purchase Amount (USD)" , color='yellow')
plt.xticks(rotation=90)  
plt.tight_layout()
plt.show()
```

```python
# ViolinPlot
sns.violinplot(data=df , x = "Payment Method" , y = "Item Purchased")
plt.tight_layout()
```

```python
# CatPlot
sns.catplot(data=df , x = "Payment Method" , y = "Item Purchased " , kind = "bar")
```

```python
# StripPlot
sns.stripplot(data = df , x = "Payment Method" , y =  "Item Purchased")
```

```python
# SwarmPlot
sns.swarmplot(data = df , x = "Payment Method" , y =  "Item Purchased")
```
